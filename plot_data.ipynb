```python
import numpy as np
import matplotlib.pyplot as plt
from data_manager import DataManager
```

```python
molecule_path = "../data/molecule/"
gold_path = "../data/gold/"

data_manager = DataManager(molecule_path=molecule_path,
                           gold_path=gold_path)

molecule_data, gold_data = data_manager.load_data(n_files=10)
```

```python
print(molecule_data.keys())
print(molecule_data["scan180226_38_2919"].keys())
print(molecule_data["scan180226_38_2919"]["metadata"])
print(molecule_data["scan180226_38_2919"]["forward_trace"].shape)
```

```python
plot_fontsize = 14
plt.figure(figsize=(12,7))
for i, key in enumerate(molecule_data):
    trace = molecule_data[key]["forward_trace"]
    plt.plot(trace[:, 0], trace[:, 1], label=key)
plt.xlabel("Time", fontsize=plot_fontsize)
plt.ylabel("Conductance", fontsize=plot_fontsize)
plt.legend()
plt.show()
```
